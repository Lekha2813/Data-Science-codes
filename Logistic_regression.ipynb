```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
bank= pd.read_csv('/content/bank-full (3).csv',delimiter=';')
```

```python
# getting dummies for the categrical variables
data=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome','month'])
data
```

```python
data.info()
```

```python
#making the binary yes or no into dummies
data.default.replace(('yes','no'),(1,0),inplace=True)
data.housing.replace(('yes','no'),(1,0),inplace=True)
data.loan.replace(('yes','no'),(1,0),inplace=True)
data.y.replace(('yes','no'),(1,0),inplace=True)
```

#Model building

```python
X=pd.concat([data.iloc[:,0:10],data.iloc[:,11:]],axis=1)
Y=pd.concat([data.iloc[:,10]])
```

```python
classifier=LogisticRegression()
classifier.fit(X,Y)
```

```python
classifier.intercept_
```

```python
classifier.coef_
```

#Model prediction

```python
y_pred=classifier.predict(X)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual':Y,
                        'predicated':y_pred})
y_pred_df
```

#Model accuracy

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score as ac

confusion_matrix=confusion_matrix(Y,y_pred)
print(confusion_matrix)
```

```python
#accuracy
ac(Y,y_pred)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```

#ROC curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds=roc_curve(Y,classifier.predict_proba(X)[:,1])
auc=roc_auc_score(Y,y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or[1-True Negative Rate]')
plt.ylabel('True Positive Rate')
```
